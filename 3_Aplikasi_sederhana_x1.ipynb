#Aplikasi menghitung luas persegi panjang

```python
panjang = int ( input("masukkan panjang dari persegi panjang (cm): "))
lebar = int(input("Masukkan lebar dari persegi panjang(cm): "))

luas = panjang*lebar

print(f"\n luas dari persegi panjang adalah {luas} cm^2.")
```

#Aplikasi menghitung volume bangun ruang

```python
sisi = int(input("masukkan sisi dari kubus (cm) : "))

volume =sisi*sisi*sisi

print(f"\n volume dari kubus adalah {volume} cm^3.")


```
